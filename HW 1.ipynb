```python
# Question 3

# randint(1,10) for random int between 1 and 10 -- 10 is inclusive in this function
# prompts user for guess repeatedly
# tells user each time if guess is lower/higher than random generated num
# correct guess -- terminate program and return # of tries/inputs to guess correctly
# counter for attempts
# while loop -- guess != randomly generated num
# handle edge cases: 0, 1, 10, 11... plus large numbers out of range(1,11) since up to but not including 11
# inner if loops to handle logic for < or > while in range
# logic for == : terminate and return 'your guess is correct' + str(attempts)
```

```python
# Question 4

from random import randint

def guessing_game():
    num = randint(1,10)
    guess = int(input('Enter an integer between 1 and 10 as a guess: '))
    attempts = 0
    
    while num != guess:
        
        if guess not in range(1,11):
            print('Guess is out of range. Please input an integer between 1 and 10.')
            guess = int(input('Enter an integer between 1 and 10 as a guess: '))
            
        elif guess < num:
            print('Your guess is too low. Try again')
            attempts += 1
            guess = int(input('Enter an integer between 1 and 10 as a guess: '))

        elif guess > num:
            print('Your guess is too high. Try again')
            attempts += 1
            guess = int(input('Enter an integer between 1 and 10 as a guess: '))
            
    attempts += 1 # when num == guess:
    print('Your guess is correct!\t Number of tries: ' + str(attempts))
    
guessing_game()
```

```python
## Question 5

# think of simplier approach -- mod % or dimensional analysis (* and //) to convert into inches and work backwords?
# take input from user in inches
# divide // by 63360 to get # of miles as an integer
# take the remainder and / by 36 to get yards
# take the new remainder and / by 12 to get feet
# take the newer remainder and what is left is the residual in inches (keep units consistent)
# handle printing only nonzero units -- empty string
# if statements with m, y, f, inch !=0: add each nonzero unit to the empty string
# print string and test input of 63360 to ensure output is only '1 miles', etc.
```

```python
# Question 6

inches = int(input('Enter a number of inches as an integer: '))   
m = int(inches // 63360) # convert from inches to miles, with no remainder
y = int((inches - m * 63360) / 36) # convert the difference from inches to yards
f = int((inches - m * 63360 - y * 36) / 12) # convert the difference from inches to feet
inch = int(inches - m * 63360 - y * 36 - f * 12) # what's left over with everything in same units [inches]
a = ''
if m != 0:
    a += str(m) + ' miles\n' 
if y != 0:
    a += str(y) + ' yards\n'
if f != 0:
    a += str(f) + ' feet\n'
if inch != 0:
    a += str(inch) + ' inches'
print(a)
```
