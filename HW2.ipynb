```python
# Question 1

new_lst = []
def only_pos_num(lst):
    for i in lst:
        if type(i) == str:
            try:
                if '.' in i:
                    i = abs(float(i))
                    new_lst.append(i)
                else:
                    i = abs(int(i))
                    new_lst.append(i)
            except:
                pass   
        elif i < 0:
            i = abs(i)
            new_lst.append(i)
        else:
            new_lst.append(i)   
    return new_lst 

print('Input:',['wow', -2.2, -4, '-3', 'hi', '-6.2', '4'], '\n', 'Output:', only_pos_num(['wow', -2.2, -4, '-3', 'hi', '-6.2', '4']))
# print('Input:',[1, 2.0, 'cat', '-0.1'], '\n', 'Output:', only_pos_num([1, 2.0, 'cat', '-0.1']))
# print('Input:',['hi', 'okay', 'DSCI 510', 'Python is cool, man'], '\n', 'Output:', only_pos_num(['hi', 'okay', 'DSCI 510', 'Python is cool, man']))
# print('Input:',[-1.1, 2, 3.54, -4, -5, -6], '\n', 'Output:', only_pos_num([-1.1, 2, 3.54, -4, -5, -6]))
```

```python
# Question 2

def read_csv(filename):
    fhand = open(filename, 'r', encoding = 'utf-8-sig')
    d = {}
    l = []
    
    for line in fhand:
        line = line.rstrip().split(',')
        l.append(line)
        
    for i in range(len(l[0])):
        c = []
        for j in range(1, len(l)):
            c.append(l[j][i])
        d[l[0][i]] = c
        
    return d

print(read_csv('Students_Q2Q3.csv'))
```

```python
# Question 3

def read_csv(filename):
    fhand = open(filename, 'r', encoding = 'utf-8-sig')
    d = {}
    l = []
    
    for line in fhand:
        line = line.rstrip().split(',')
        l.append(line)
        
    for i in range(len(l[0])):
        c = []
        for j in range(1, len(l)):
            c.append(l[j][i])
        d[l[0][i]] = c
        
    return d

def transform_csv(filename): 
    a = read_csv(filename)
    d2 = {}
    lst2 = []

    for k, v in a.items():
        v.reverse()
        v.append(k)
        lst2.append(v)

    for i in range(len(lst2)):
        c2 = []
        for j in range(1, len(lst2[0])):
            c2.append(lst2[i][j])
        c2.reverse()
        d2[lst2[i][0]] = c2
        
    return d2

print(transform_csv('Students_Q2Q3.csv'))
```
