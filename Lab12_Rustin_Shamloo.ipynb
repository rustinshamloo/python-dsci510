Complete the `feed_data` function which will feed all data from `filename` (argument) into a sqlite database. While doing this, make sure of the following: 
- Don't include the first row(it has the name of the columns in it.)
- Database name: lab12.db
- Table name: `table_name` (argument)
- Columns: date(type: DATE) and price(type: REAL)

For this function, you don't have to print/return anything. Do not change given code.   


Note 1: Use Lab6Code for easier csv manipulation.  
Note 2: Date column in the file will have "YYYY-MM-DD" format.   
Note 3: Use the datetime module for date manipulation(Refer: https://docs.python.org/3/library/datetime.html (Specifically, look at this: https://docs.python.org/3/library/datetime.html#datetime.datetime.strptime))  
Note 4: Once you use strptime, you will get datetime. But, you need to insert date. (See `date()` function in datetime)  
Note 5: In the print output, if you see a string instead of datetime.date(...), then you do not have the data stored in the correct format.  

---
Grading Rubric:  
+5 points if data is inserted and table exists.  
+10 points if data is stored in correct format. 

```python
import csv
import Lab6Code
import sqlite3
import datetime

def feed_data(filename, table_name):
    conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
    cur = conn.cursor()
    
    data = Lab6Code.csv_dataset(filename, ",")
    cur.execute('DROP TABLE IF EXISTS ' + table_name + '')
    cur.execute('CREATE TABLE ' + table_name + 
               ' (date DATE, price REAL)')
    
    for row in range(1, len(data)): 
        item = data[row]
        cur.execute('INSERT INTO ' + table_name + ' VALUES(?,?)', item)
        
    cur.execute('SELECT * FROM ' + table_name + ' LIMIT 1')
    return_val = cur.fetchall()
    print(return_val)

    conn.commit()
    conn.close()
    
feed_data('brent-daily_csv.csv', 'brent_oil_price')
feed_data('wti-daily_csv.csv', 'wti_oil_price')
```

For this question, you have to write queries on the `lab12.db` for the following questions and print the results of the query.  
1. Get brent oil prices for year 2020. Print format : [(date, price)]   
2. Number of days for which there exists prices for both (brent and wti) oil. Print format: int 
3. Get brent oil prices for dates when brent oil price is higher than wti oil prices. Print format: [(date, price)]
4. Find maximum occurance of a brent oil price. Print format: int
5. Get information of brent and wti oil price which is sorted by difference between brent and wti oil price. Print Format: [(date, brent oil price, wti oil price)]

---
Rubric: 3 points for each query

```python
#Get brent oil prices for year 2020. Print format : [(date, price)]    

conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
cur = conn.cursor()

a = cur.execute("SELECT * FROM 'brent_oil_price' WHERE strftime('%Y', date)='2020'").fetchall()
print(a)

conn.commit()
conn.close()
```

```python
#Number of days for which there exists prices for both brent and wti oil. Print format: int

conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
cur = conn.cursor()

b = cur.execute("SELECT COUNT(*) FROM brent_oil_price b, wti_oil_price w WHERE b.Date = w.Date").fetchone()[0]
print(b)

conn.commit()
conn.close()
```

```python
#Get brent oil prices for dates when brent oil price is higher than wti oil prices. Print format: [(date, price)]     

conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
cur = conn.cursor()

c = cur.execute('''SELECT b.Date, b.Price FROM brent_oil_price b, wti_oil_price w 
                WHERE b.Date = w.Date AND b.Price > w.Price''').fetchall()
print(c)

conn.commit()
conn.close()
```

```python
#Find maximum occurance of a brent oil price. Print format: int

conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
cur = conn.cursor()

d = cur.execute('''SELECT COUNT(b.Price) AS cnt FROM brent_oil_price b 
                GROUP BY b.Price ORDER BY cnt DESC LIMIT 1''').fetchone()[0]
print(d)

conn.commit()
conn.close()
```

```python
#Get information of brent and wti oil price which is sorted by difference between brent and wti oil price. 
#Print Format: [(date, brent oil price, wti oil price)]

conn = sqlite3.connect('lab12.db', detect_types=sqlite3.PARSE_DECLTYPES|sqlite3.PARSE_COLNAMES)
cur = conn.cursor()

cur.execute('''SELECT b.Date, b.Price, w.Price FROM brent_oil_price b 
               INNER JOIN wti_oil_price w ON w.Date = b.Date ORDER BY (w.Price - b.Price)''')

print(cur.fetchall())

conn.commit()
conn.close()
```
